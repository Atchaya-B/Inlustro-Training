```python
df=df.dropna()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Survived",data=df,palette="pastel")
plt.title("Survived count",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x="Sex",y="Survived",data=df,palette="mako")
plt.title("Survived count",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"],bins=30,kde=True,color="blue")
plt.title("Age Distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[["Sex","Fare"]],palette="coolwarm")
plt.title("Boxplot for outliers determined",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age",y="Fare",data=df,palette="muted")
plt.title("Age vs Fare",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(x="Pclass",y="Age",data=df,palette="pastel")
plt.title("class vs Age",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df,x="Fare",color="red")
plt.title("Fare Distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x="PassengerId",y="Fare",data=df)
plt.title("PassengerId vs Fare",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.pie(x="Pclass",data=df)
plt.title("Pclass distribution")
plt.show()
```
